# Extracting a Data from a website by statically providing url list with Python library urllib and Beautiful Soup.

#### Here i am using a url from Flipkart which is a  famous Ecommerce platform in india. The entire exercise is conducted for the educational parpose only. This is the most basic version of scraping any web pages.

```python
# First we need to import the library which we are going to be using

from urllib.request import urlopen as uopen
from bs4 import BeautifulSoup as soup
```

```python
# Here i am statically providing the urls to be used for Scraping the information like brand name,item name and cost.

urls = ["https://www.flipkart.com/search?q=memory+card&otracker=start&as-show=off&as=off",
       "https://www.flipkart.com/search?as=off&as-show=off&otracker=start&page=2&q=memory+card&viewType=grid",
       "https://www.flipkart.com/search?as=off&as-show=off&otracker=start&page=3&q=memory+card&viewType=grid",
       "https://www.flipkart.com/search?as=off&as-show=off&otracker=start&page=4&q=memory+card&viewType=grid"]
```

```python
# Here is the python script for creating the csv file and adding the header into it.

header = "brand,item_name,current_price,\n"
filename = "./product1.csv"
with open(filename, "w") as f:
    f.write(header)

# Below python script for scraping the data from the web.
        
for url in urls:
    data = uopen(url)
    page = data.read()
    soup_page = soup(page,"html.parser")
    data.close()

    items_name = soup_page.find_all("a",{"class":"_2cLu-l"})
    item_title = []
    for item in items_name:
        item_title.append(item.get("title"))


    brand = []
    for item in item_title:
        brand.append(item.split()[0])

    items_current_price = soup_page.find_all("div",{"class":"_1vC4OE"})
    current_price = []
    for p in items_current_price:
        current_price.append(p.text)
    
    for cp in current_price:
        print(type(cp))
        cp = cp.replace(',','')
    
    for b,i,p in zip(brand,item_title,current_price):
        print("brand: ",b)
        print("item_title: ",i)
        print("price: ",p)
        print()

# Finally we need to update the csv file we just created with the data we scraped already.

    filename = "./product1.csv"
    with open(filename, "a") as f:
        for i in range(len(brand)):
            current_price[i] = current_price[i].replace(',','')
            f.write(brand[i]+","+item_title[i]+","+current_price[i].replace('₹','')+"\n")
        f.close()
```
